```python
from transformers import AutoModelForCausalLM, AutoTokenizer

model_name = "meta-llama/Llama-3.2-3B-Instruct" # Or "meta-llama/Llama-3.2-3B" for the base model

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# Now you can use the tokenizer and model for tasks like text generation
```

```python
prompt =  "Narendra Modi is the Prime Minister of India. He is a"

# Encode the prompt
input_ids = tokenizer.encode(prompt, return_tensors="pt")

# Generate text
output = model.generate(input_ids,
                        max_length=100,          # Maximum length of the generated text
                        num_return_sequences=1, # Generate one sequence
                        temperature=0.9,        # Adjust for creativity (higher is more creative)
                        top_p=0.9)              # Another sampling parameter

# Decode the generated output
generated_text = tokenizer.decode(output[0], skip_special_tokens=True)

print(generated_text)
```
